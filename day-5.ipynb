## Dessert inheritance

```python
class Dessert:
    def __init__(self, name, calories):
        self.name = name
        self.calories = calories

    def healthy(self):
        if self.calories < 200:
            return True

    def delicious(self):
        return True

class JellyBean(Dessert):
    def __init__(self, name, calories, flavor):
        super().__init__(name, calories)
        self.flavor = flavor

    def delicious(self):
        if self.flavor != 'black licorice':
            return super().delicious()
        else:
            return False
```

## Sum of digits

```python
def digital_root(n):
    n = str(n)

    new_list = []
    for i in n:
        if len(n) > 1:
            new_list.append(int(i))
        else:
            return int(n)

    # print(new_list)

    sum_new_list = str(sum(new_list))

    if len(sum_new_list) > 1:
        return digital_root(sum(new_list))
    else:
        return sum(new_list)
```

## Stop spinning my words

```python
def spin_words(sentence):

    different_words = sentence.split()
    # print(different_words)

    def return_one_word(word):
        new_list = []
        for i in range(1,len(word)+1):
            new_list.append(word[-i])

        new_sentence = str(new_list).replace('[', '').replace(']', '').replace(',','').replace("'",'').replace(' ',"")
        return f'{new_sentence}'

    if len(different_words) != 1:
        new_string = str()
        for i in range(len(different_words)):
            if len(different_words[i]) > 4 :
                new_string = new_string + (str(return_one_word(different_words[i]))) + str(' ')
            else :
                new_string = new_string + different_words[i] + str(' ')

        return (new_string)[:-1]

    else:
        if len(different_words[0]) > 4 :
            return return_one_word(different_words[0])
        else:
            return sentence
```

## Create a phone number

```python
def create_phone_number(n):
    new_string = str("(")

    for i in n[:3]:
        new_string = new_string + str(i)

    new_string = new_string + str(') ')

    for i in n[3:6]:
        new_string = new_string + str(i)

    new_string = new_string + str('-')

    for i in n[6:]:
        new_string = new_string + str(i)


    return new_string
```
